```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
import os 
import nltk
import time
from nltk.corpus import stopwords
nltk.download('stopwords')

path_theo = '/Users/theojanson/Desktop/movie_reviews/'
path = path_theo

```

```python
def create_sets(folder, kind):
  data = []
  labels = []
  os.chdir(path+kind+"/"+folder)
  for t in os.listdir(): #iterate over every file in directory
    if folder[0].lower() == 'p':
      labels.append(1)  #append label
    else: 
      labels.append(0)                
    with open(t) as f:
      data.append(f.readlines()[0]) #read file
  return pd.DataFrame({'Data': data, 'Labels': labels}) #create dataframe with the lists

def merge_sets(kind):
    positives = create_sets('pos', kind)
    negatives = create_sets('neg', kind)
    return positives.append(negatives, ignore_index=True)
    
path_save = 'Users/theojanson/McGill/COMP 551/Code/Assignment 2/'
merge_sets('train').to_csv(path_save+'train_set.csv')

```
